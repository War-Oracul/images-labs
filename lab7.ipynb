```python
import cv2
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

#     
source_image = cv2.imread('army.jpg')
fragment_image = cv2.imread('chel.jpg')

#    
result = cv2.matchTemplate(source_image, fragment_image, cv2.TM_CCOEFF_NORMED)

#     
threshold = 0.8
locations = np.where(result >= threshold)

#     
for loc in zip(*locations[::-1]):
    cv2.rectangle(source_image, loc, (loc[0] + fragment_image.shape[1], loc[1] + fragment_image.shape[0]), (0, 255, 0), 2)

plt.figure(figsize=(15,20))
plt.subplot(1, 2, 1)
plt.imshow(cv.cvtColor(source_image, cv.COLOR_BGR2RGB))
plt.show()
```
